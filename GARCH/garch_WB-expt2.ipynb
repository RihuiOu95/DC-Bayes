### Divide-and-conquer:

```python
m = 10
tstarts = np.arange(m).astype(int)
tends = 1 + tstarts
tstarts *= int(T/m)
tends *= int(T/m)
```

```python
for i in range(m) :
    tstart, tend = tstarts[i], tends[i]
    data = dict(T=tend-tstart, p=p, q=q, r=max(p,q), d=d, y=y[tstart:tend], X=X[:,tstart:tend], power=T/(tend-tstart))
    
    fit = sm.sampling(data=data, thin=1, n_jobs=min(10,n_chains), chains=n_chains, init="random", iter=n_iter)
    file = open("mcmc_wb_chunk"+str(i+1)+".pkl", "wb")
    pickle.dump(fit.extract(), file)
    file.close()
    
    mle = sm.optimizing(data=data)
    file = open("mle_chunk"+str(i+1)+".pkl", "wb")
    pickle.dump(mle, file)
    file.close()
```

# Laplace Approx

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import torch.nn.functional as F
```

```python
class GARCH(nn.Module):
    def __init__(self, p, q):
        super(GARCH, self).__init__()
        self.p = p
        self.q = q
        self.raw_omega = nn.Parameter(-torch.randn(1))
        self.raw_alpha = nn.Parameter(-10 * torch.randn(p))
        self.raw_beta = nn.Parameter(-10 * torch.randn(q))

    def forward(self, returns):
        T = returns.shape[0]
        sigma2 = torch.zeros(T)
        omega = F.softplus(self.raw_omega)
        alpha = F.softplus(self.raw_alpha)
        beta = F.softplus(self.raw_beta)
        
        for t in range(max(self.p, self.q)):
            sigma2[t] = returns[:self.p].pow(2).mean()

        for t in range(max(self.p, self.q), T):
            alpha_term = torch.sum(alpha * returns[max(0, t-self.p):t].pow(2).clone())
            beta_term = torch.sum(beta * sigma2[max(0, t-self.q):t].clone())
            sigma2[t] = omega + alpha_term + beta_term

        return sigma2

# Negative log-likelihood loss
def nll_loss(sigma2, returns):
    T = len(returns)
    log_likelihood = -(torch.log(sigma2) + returns.pow(2) / sigma2)
    return -torch.sum(log_likelihood) 

# Hyperparameters
p, q = 2, 2
learning_rate = 0.02
num_epochs = 300

# Sample data
returns = torch.tensor(y).reshape(len(y),1)

# Model and optimizer
model = GARCH(p, q)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```
